```python
import pandas as pd
```

```python
df1 = pd.read_csv('Originales/Starlust.csv')
df2 = pd.read_csv("dataset1.csv")
```

```python
df1.head()
```

```python
#Quiero dejar solo las columnas M y SIZE
df1 = df1[['M', 'SIZE']]
df1.head(10)
```

```python
for i in df1["SIZE"]:
    size = i.split("x")
    print(size)
    a = size[0]
    if (len(size) == 1) or (len(size) == 2 and size[1] == ''):
        c = round(float(a)**2, 2)
    else:
        b = size[1]
        c = round(float(a)*float(b), 2)
    df1["SIZE"] = df1["SIZE"].replace(i, c)
```

```python
df2.head()
```

```python
df = pd.merge(df2, df1, left_on="Messier", right_on="M", how="inner")
```

```python
#eliminar la columna dimensions y m
df = df.drop(columns=["Dimensions", "M"])
#renombrar la columna size por dimensions
df = df.rename(columns={"SIZE": "Dimensions"})
```
